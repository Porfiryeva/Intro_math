### 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы


```python
# окружность
r = 5
x = np.linspace(-r, r, 1000)  # центр в точке (0, 0), размах x от -r до r
y1 = [math.sqrt(r**2 - el**2) for el in x]
y2 = [-math.sqrt(abs(r**2 - el**2)) for el in x]

plt.figure(figsize=(6,6))  # квадратная область построения
plt.plot(x, y1)
plt.plot(x, y2)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# эллипс
a = 5
b = 3
x = np.linspace(-a, a, 1000)
y1 = [math.sqrt(b**2 - b**2 * el**2 / a**2) for el in x]
y2 = [-math.sqrt(b**2 - b**2 * el**2 / a**2) for el in x]

plt.figure(figsize=(6,6))
plt.plot(x, y1)
plt.plot(x, y2)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# гипербола
a = 3
b = 2
x1 = np.linspace(-10, -a, 1000)
x2 = np.linspace(a, 10, 1000)
# левая верхняя ветвь
y1 = [math.sqrt(b**2 * el**2 / a**2 - b**2 ) for el in x1]
# левая нижняя ветвь
y1_1 = [-math.sqrt(b**2 * el**2 / a**2 - b**2 ) for el in x1]
y2 = [math.sqrt(b**2 * el**2 / a**2 - b**2) for el in x2]
y2_2 = [-math.sqrt(b**2 * el**2 / a**2 - b**2) for el in x2]

plt.figure(figsize=(6,6))
plt.plot(x1, y1)
plt.plot(x1, y1_1)
plt.plot(x2, y2)
plt.plot(x2, y2_2)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 1. Задание (в программе)
Нарисуйте график функции: для некоторых (2-3 различных) значений параметров k, a, b: $$y = k*cos(x – a) + b$$

```python
def f(x, k=1, a=0, b=0):
    return k*np.sin(x - a) + b

x = np.linspace(-2*np.pi, 3*np.pi, 201)
plt.plot(x, f(x=x, a=1), label='sin(x-1)')  # синус сдвигается -> на 1
plt.plot(x, f(x=x, k=2), label='2sin(x)')  # синус растягивается в 2 раза по oy
plt.plot(x, f(x=x, b=-1), label='sin(x)-1')  # синус сдвигается вниз на 1
plt.plot(x, f(x=2*x, k=1.5, a = -3, b = 2), label='1.5sin(2x+3)+2')  # всё сразу
plt.plot([-2*np.pi, 3*np.pi],[0,0], color ='black', linewidth=0.75, linestyle ="--")
plt.plot([0, 0],[-2, 3.5], color ='black', linewidth=0.75, linestyle ="--")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### 4. Решите систему уравнений:

    exp(x) + x∙(1 – y) = 1
    y = x2 – 1

```python
# график для поиска точек старта
x = np.linspace(-2, 5, 2000)
plt.figure(figsize=(6,6))
plt.plot(x, (np.exp(x) - 1) / x + 1)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,20) 
plt.grid(True)
plt.show()
```

Исходя из графика - имеет смысл проверить три точки (-1.5, 2), (2.5, 5.5), (4, 16)

```python
def equations(param):
    x, y = param
    return math.exp(x) + x * (1 - y) - 1, x**2 - 1 - y

x1, y1 = fsolve(equations, (-1.5, 2))
print(x1, y1)
x2, y2 = fsolve(equations, (2.5, 5.5))
print(x2, y2)
x3, y3 = fsolve(equations, (4, 16))
print(x3, y3)
```

Решите систему неравенств:

```python
x = np.linspace(-2, 5, 500)
plt.figure(figsize=(6,6))
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) - 1) / x + 1)
plt.xlabel('x')
plt.ylabel('y')
# plt.ylim(-2,20) 
plt.grid(True)
plt.show()
```
